```python
# Imports Libraries
import gradio as gr
import numpy as np
import tensorflow as tf
from PIL import Image
```

```python
# Load model
model = tf.keras.models.load_model("/content/insects_model.keras")
class_names = np.load("/content/class_names.npy", allow_pickle=True)
```

```python
# Inference function
def classify_image(image):
    image = image.resize((224, 224))
    image = np.array(image).astype('float32') / 255.0
    image = np.expand_dims(image, axis=0)
    preds = model.predict(image)[0]
    return {class_names[i]: float(preds[i]) for i in range(len(class_names))}
```

```python
# Launch UI
gr.Interface(
    fn=classify_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Label(num_top_classes=1),
    title="Farm Insects Prediction",
    description="Upload an image of an insects to identify its class."
).launch(share=True)
```
